```python
# Load packages
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
##1# defining the url for the site
base_site = 'https://www.ikea.com/eg/en/new/new-products/'

##2# making a get request
response = requests.get(base_site)

##3# Extracting the content of HTML
html = response.content

##4# convert HTML to a BeautifulSoup object. This step will allow you to parse out content from the HTML more easily
# choosing the proper parser
soup = BeautifulSoup(html, 'lxml')

##5# Extracting products and prices
brand = soup.find_all('div', {'class':'range-revamp-header-section__title--small notranslate'})
products = soup.find_all('span', {'class' : "range-revamp-header-section__description-text"})
prices = soup.find_all('div', {'class' : 'range-revamp-compact-price-package__price-wrapper'})
#prices
```

```python
##6# Extract product names and prices
brand_name = [brand.string for brand in brand]
products_name = [product.string for product in products]
products_prices = [price.find('span', {'class' : 'range-revamp-price__integer'}).string for price in prices] 
print(len(brand_name))
print(len(products_name))
print(len(products_prices))
```

```python
IKEA_products = pd.DataFrame()

IKEA_products['Brand'] = brand_name
IKEA_products['Products'] = products_name
IKEA_products['Prices'] = products_prices
IKEA_products
```

```python
IKEA_products.to_csv("IKEA_product.csv", index = False, header =True)
```
